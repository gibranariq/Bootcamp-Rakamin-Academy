```python
# # Buat kolom tahun cara 1
# tahun = df['Order Date'].dt.year

# # add a new column 'Year' to the DataFrame
# df['Year'] = tahun

# Buat kolom tahun cara 2
# Ensure 'Order Date' is in datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Create 'Year' column using lambda
tahun = df['Order Date'].apply(lambda x: str(x)[:4])
tahun = tahun.astype(int)  

# buat kolom baru
df['Year'] = tahun
df
```

```python
#filter for South region in 2016
sales_2016 = df[(df['Region']== "South") & (df['Year'] == 2016)].Sales.sum()
profit_2016 = df[(df['Region']== "South") & (df['Year'] == 2016)].Profit.sum()

print('Jumlah sales pada region South pada tahun 2016 adalah', round(sales_2016))
print('Jumlah profit pada region South pada tahun 2016 adalah', round(profit_2016))
```

```python
#filter for West region in 2017
sales_2017 = df[(df['Region']== "West") & (df['Year'] == 2017)].Sales.sum()
profit_2017 = df[(df['Region']== "West") & (df['Year'] == 2017)].Profit.sum()

#hasil
print('Jumlah sales pada region West pada tahun 2017 adalah', round(sales_2017))
print('Jumlah profit pada region West pada tahun 2017 adalah', round(profit_2017))
```

```python
def jumlah_sales_dan_profit(Region, Year=2017):
    total_sales = df[(df['Region']== Region) & (df['Year'] == 2017)].Sales.sum()
    total_profit = df[(df['Region']== Region) & (df['Year'] == 2017)].Profit.sum()

    print(f"Total sales di region {Region} adalah {total_sales}")
    print(f"Total profit di region {Region} adalah {total_profit}")
jumlah_sales_dan_profit("West")
```

```python
def count_sales_profit(region, year=2017):
  new_df = df[(df['Year'] == year) & (df['Region'] == region.capitalize())]
  return new_df.agg({'Sales' : 'sum','Profit' : 'sum'})

print('Soal 1')
south_2016 = count_sales_profit('south', 2016)
print(f"Total sales dan profit di South Region tahun 2016:\n"
      f"Sales: {south_2016['Sales']:.2f}, Profit: {south_2016['Profit']:.2f}")

print('Soal 2')
west_2017 = count_sales_profit('west', 2017)
print(f"Total sales dan profit di West Region tahun 2017:\n"
      f"Sales: {west_2017['Sales']:.2f}, Profit: {west_2017['Profit']:.2f}")
```

```python
df_sales = df.groupby(['Year','Customer ID', 'Profit'])['Sales'].sum().reset_index()
df_sales
# df_sales.columns = ["Customer ID", "Total Sales"]
```

```python
df_sales_desc = df_sales.groupby(['Year'])['Sales'].describe(percentiles= [.25, .5, .75, .9, .95]).reset_index()
df_sales_desc
```

```python
df_sales_merge = df_sales.merge(df_sales_desc, on='Year', how='inner')
df_sales_merge
```

```python
def segment(x):
    if x['Sales'] < x['50%']:
        segment = 'Low'
    elif x['Sales'] <= x['75%'] and x['Sales'] >= x['50%']:
         segment = "Mid"
    else:
        segment = "High"
    return segment
```

```python
df_sales_merge['segment'] = df_sales_merge.apply(lambda x: segment(x), axis=1)
df_sales_merge.head()
```
