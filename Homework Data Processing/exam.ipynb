```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("job_data_analyst_details.csv")
df
```

```python
# Filter perusahaan di industri Video Games dengan Rating > 4.5
antoni_df = df[(df['Industry'] == 'Video Games') & (df['Rating'] > 4.5)]

# Ambil lokasi perusahaan yang memenuhi kriteria
antoni_locations = antoni_df['Location'].unique()

# Output lokasi
print("Lokasi perusahaan yang memenuhi kriteria Antoni:", antoni_locations)
```

```python
# Filter perusahaan dengan lokasi mengandung kata "Houston"
houston_df = df[df['Location'].str.contains('Houston', na=False)]

# Hitung rata-rata Rating
houston_avg_rating = houston_df['Rating'].mean()

# Output rata-rata Rating
print("Rata-rata Rating di lokasi Houston:", houston_avg_rating)
```

```python
# Filter perusahaan dengan Industry = 'Internet' dan Founded > 2015
internet_df = df[(df['Industry'] == 'Internet') & (df['Founded'] > 2015)]

# Hitung jumlah lowongan berdasarkan Job Title
job_counts = internet_df['Job Title'].value_counts()

# Ambil Job Title dengan lowongan terbanyak
most_jobs_title = job_counts.idxmax()

# Output Job Title
print("Job Title dengan lowongan terbanyak:", most_jobs_title)
```

```python
# Filter perusahaan dengan Founded = 2018
df1 = df[df['Founded'] == 2018]

# Filter perusahaan dengan Founded = 1995
df2 = df[df['Founded'] == 1995]

# Gabungkan kedua dataframe
appended_df = pd.concat([df1, df2])

# Hitung total lowongan pekerjaan
total_jobs_after_append = appended_df['Job Title'].count()

# Output total lowongan
print("Total lowongan setelah append:", total_jobs_after_append)
```

```python
# Filter perusahaan dengan kriteria yang diberikan
filtered_df = df[(df['Founded'] > 2017) & (df['Rating'] >= 3.5) & (df['Rating'] <= 4.5)]

# Tambahkan kolom rating_baru
filtered_df['rating_baru'] = filtered_df['Rating'] + 0.5

# Hitung rata-rata rating_baru
avg_rating_baru = filtered_df['rating_baru'].mean()

# Output rata-rata rating_baru
print("Rata-rata rating_baru:", avg_rating_baru)
```

```python
# Filter df1: jobs dengan description mengandung "attitude" dan Rating >= 1
df1 = df[df['Job Description'].str.contains('attitude', case=False, na=False) & (df['Rating'] >= 1)]

# Filter df2: jobs di New York, NY dan Rating >= 1
df2 = df[(df['Location'] == 'New York, NY') & (df['Rating'] >= 1)]

# Gabungkan kedua dataframe
appended_df = pd.concat([df1, df2])

# Hitung rata-rata Rating
mean_rating = appended_df['Rating'].mean()

# Output rata-rata Rating
print("Mean Rating dari seluruh jobs yang tersedia:", mean_rating)
```

```python
# Filter lokasi Lake Success, NY atau Los Angeles, CA
filtered_df = df[(df['Location'].isin(['Lake Success, NY', 'Los Angeles, CA']))]

# Filter jobs yang mensyaratkan SQL
sql_jobs_df = filtered_df[filtered_df['Job Description'].str.contains('SQL', case=False, na=False)]

# Hitung total lowongan pekerjaan
total_sql_jobs = sql_jobs_df['Job Title'].count()

# Output total lowongan pekerjaan
print("Total lowongan pekerjaan yang mensyaratkan SQL:", total_sql_jobs)
```
